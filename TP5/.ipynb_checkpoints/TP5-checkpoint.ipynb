&emsp;$1.$&emsp;$A)$&emsp; La frecuencia de muestreo del ADC debe ser $f_s \geq 20kHz$ siendo esta 2 veces la frecuencia máxima a la que se transporta información. De todas formas, para estar más holgado, elijo una $f_s=40kHz$

$B)$ &emsp; Elijo una atenuación en la banda de detenida de $20dB$, de tal manera que las componentes de frecuencia por encima de los $10kHz$ queden por debajo del piso de ruido de la señal, ya que atenuar más que eso sería innecesario.

```python
# Inicialización e importación de módulos

# Módulos para Jupyter
import warnings
warnings.filterwarnings('ignore')

# Módulos importantantes
import scipy.signal as sig
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

fig_sz_x = 10
fig_sz_y = 7
fig_dpi = 100 # dpi

fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x,fig_sz_y)
plt.rcParams.update({'font.size':fig_font_size})


fs = 40*(10**3) # Hz
nyq_frec = fs / 2

# Plantilla

# filter design
ripple = 1 # dB
atenuacion = 20 # dB

ws = 11*(10**3) #Hz
wp = 10*(10**3) #Hz

wsn = ws/nyq_frec
wpn = wp/nyq_frec

lp_sos_butter = sig.iirdesign(wpn, wsn, ripple, atenuacion, False, 'butter', 'sos')

cant_coeficientes = 1501

Omega, H = sig.sosfreqz(lp_sos_butter, worN=2048, whole=False, fs=fs)

plt.figure()

plt.plot(Omega, 20 * np.log10(np.abs(H)), label='IIR')

plt.title('Filtro diseñado')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Módulo [dB]')
plt.grid()
plt.axis([0, 12500, -60, 5 ]);

plt.figure()

plt.plot(Omega, 20 * np.log10(np.abs(H)), label='IIR')

plt.title('Filtro diseñado')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Módulo [dB]')
plt.grid()
```
