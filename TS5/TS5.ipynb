$$Par[F(s)]=M(s)=\frac{m_1m_2-n_1n_2}{m_2^2-n_2^2}=\frac{-1-s^2}{1-s^2}$$

$$Impar[F(s)]=N(s)=\frac{n_1m_2-m_1n_2}{m_2^2-n_2^2}=\frac{s+s}{1-s^2}=\frac{2s}{1-s^2}$$

$$Real\{F(j\omega)\}=M(j\omega)=\frac{-(j\omega)^2-1}{1-(j\omega)^2}=\frac{\omega^2-1}{\omega^2+1}$$

$$Img\{F(j\omega)\}=N(j\omega)=\frac{2(j\omega)}{1-(j\omega)^2}=j\frac{2\omega}{\omega^2+1}$$

$$jtan(\phi(j\omega))=\frac{Img\{F(j\omega)\}}{Real\{F(j\omega)\}}=j\frac{2\omega}{\omega^2-1}$$

$$\phi(j\omega)=tan^{-1}(\frac{2\omega}{\omega^2-1})\;\;\;\;\;\;\;\;\;\;\;\tau(j\omega)=-\frac{d\phi}{d\omega}$$

$$\tau(j\omega)=-\frac{1}{1+[\frac{2\omega}{\omega^2-1}]^2}\cdot(-2)\cdot\frac{(\omega^2 + 1)}{(\omega^2 - 1)^2}
=\frac{2\cdot(\omega^2 + 1)}{[\omega^2-1]^2+2\omega^2}$$

$$\tau(j\omega)=\frac{2\cdot(\omega^2 + 1)}{\omega^4-2\omega^2+1+2\omega^2}=\frac{2\cdot(\omega^2 + 1)}{\omega^4+1}$$

$$Cálculo\;retardo\;de\;grupo\;\;\;\;\;\;\;\;\;\;\;\;\;\;\tau(j\omega)=\frac{2\cdot(\omega^2+1)}{\omega^4+1}$$

$$\phi(j\omega)=\frac{\pi}{2}-arctg(\frac{6\omega}{-\omega^2+4})\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
\phi_z=\frac{\pi}{2}\rightarrow Cero\;en\;el\;origen\;\;\;\;\;\;\;\;\;\;\phi_p=arctg(\frac{6\omega}{-\omega^2+4})$$

$$\phi_p=arctg(\frac{Img\{Q(\omega)\}}{Real\{Q(\omega)\}})\rightarrow Q(\omega)=-\omega^2+6j\omega+4
\rightarrow Q(s)=s^2+6s+4$$

$$F(s)=\frac{6s}{s^2+6s+4}$$

&emsp;$b)$&emsp;

$$\omega=0\;\;\;\;\;\;\;\;\;\phi(\omega)=\frac{\pi}{2}-tan^{-1}(\frac{6\omega}{-\omega^2+4})
\rightarrow\phi(0)=\frac{\pi}{2}-tan^{-1}(\frac{0}{4})
\rightarrow\phi(0)=\frac{\pi}{2}$$

$$\omega \rightarrow \infty \;\;\;\;\;\;\;\;\;\phi(\omega)=\frac{\pi}{2}-tan^{-1}(\frac{6\omega}{-\omega^2+4})
\rightarrow\phi(\omega)=\lim_{\omega \to \infty}[\frac{\pi}{2}-tan^{-1}(\frac{6}{-\omega\cdot(1+\frac{4}{\omega^2})})]$$

$$\phi(\omega)=\frac{\pi}{2}-tan^{-1}(\frac{6}{\infty}))=\frac{\pi}{2}-tan^{-1}(0)=\frac{\pi}{2}$$
